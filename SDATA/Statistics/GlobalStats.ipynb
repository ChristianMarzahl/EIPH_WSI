# Per Image analysis

## Human

```python
data = []

temp = density_annotations[density_annotations["species"]=="Human"]
for image_name in temp["image_name"].unique():
    
    imageDf = temp[temp["image_name"]==image_name]
    
    _0 = len(imageDf[imageDf["grade"]==0])
    _1 = len(imageDf[imageDf["grade"]==1])
    _2 = len(imageDf[imageDf["grade"]==2])
    _3 = len(imageDf[imageDf["grade"]==3])
    _4 = len(imageDf[imageDf["grade"]==4])
    
    
    data.append([image_name.replace("tiff","svs"), _0, _1, _2, _3, _4, sum([_0,_1,_2,_3,_4])])
    
data = pd.DataFrame(data, columns=["image_name", "_0", "_1", "_2" ,"_3", "_4", "sum"])
data.sort_values(by=['sum'], ascending=False)
```
